```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[0.25,0.33,0.5,1,2]
y=[75.28,76.19,77.87,91.34,92.22]
```

```python
plt.figure(figsize=(12,6))
plt.plot(x,y,marker='o')
plt.xlabel('Transmission rate(Hz)')
plt.ylabel('Average current(mA)')
plt.title('Average current vs Transmission rate')
# plt.grid()
# plt.legend()
plt.show()
```
